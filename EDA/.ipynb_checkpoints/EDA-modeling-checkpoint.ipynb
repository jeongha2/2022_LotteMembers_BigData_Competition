```python
def data_partition_with_valid(fname):
    """
    Split Dataset
    """
    usernum = 0
    itemnum = 0
    User = defaultdict(list)
    user_train = {}
    user_valid = {}
    user_test = {}
    
    print('Preparing data...')
    f = open('%s.txt' % fname, 'r')
    time_set = set()

    user_count = defaultdict(int)
    item_count = defaultdict(int)
    for line in f:
        try:
            u, i, rating, timestamp, buy_am, clac_hlv_nm, clac_mcls_nm, cop_c, chnl_dv, de_dt_month, ma_fem_dv, ages, zon_hlv = line.rstrip().split('\t')
        except:
            u, i, timestamp, buy_am, clac_hlv_nm, clac_mcls_nm, cop_c, chnl_dv, de_dt_month, ma_fem_dv, ages, zon_hlv = line.rstrip().split('\t')
        u = int(u)
        i = int(i)
        user_count[u]+=1
        item_count[i]+=1
    f.close()
    f = open('%s.txt' % fname, 'r') # try?...ugly data pre-processing code
    for line in f:
        try:
            u, i, rating, timestamp, buy_am, clac_hlv_nm, clac_mcls_nm, cop_c, chnl_dv, de_dt_month, ma_fem_dv, ages, zon_hlv = line.rstrip().split('\t')
        except:
            u, i, timestamp, buy_am, clac_hlv_nm, clac_mcls_nm, cop_c, chnl_dv, de_dt_month, ma_fem_dv, ages, zon_hlv = line.rstrip().split('\t')
        u = int(u)
        i = int(i)
        timestamp = float(timestamp)
        buy_am = float(buy_am)
        clac_hlv_nm = int(clac_hlv_nm)
        clac_mcls_nm = int(clac_mcls_nm)
        cop_c = int(cop_c)
        chnl_dv = int(chnl_dv)
        de_dt_month = int(de_dt_month)
        ma_fem_dv = int(ma_fem_dv)
        ages = int(ages)
        zon_hlv = int(zon_hlv)
        if user_count[u]<5 or item_count[i]<5: # hard-coded
            continue
        time_set.add(timestamp)
        User[u].append([i, timestamp, buy_am, clac_hlv_nm, clac_mcls_nm, cop_c, chnl_dv, de_dt_month, ma_fem_dv, ages, zon_hlv])
    f.close()
    time_map = timeSlice(time_set)
    User, usernum, itemnum, timenum = cleanAndsort(User, time_map)

    for user in User:
        nfeedback = len(User[user])
        if nfeedback < 3:
            user_train[user] = User[user]
            user_valid[user] = []
            user_test[user] = []
        else:
            user_train[user] = User[user][:-1]
            user_valid[user] = []
            user_valid[user].append(User[user][-2])
            user_test[user] = []
            user_test[user].append(User[user][-1])
    print('Preparing done...')
    return [user_train, user_valid, user_test, usernum, itemnum, timenum]
```

```python
seqs = model.item_emb(log_seqs)
seqs *= model.item_emb.embedding_dim ** 0.5
seqs = model.item_emb_dropout(seqs)

positions = torch.LongTensor(np.tile(np.array(range(seqs.shape[1])), [seqs.shape[0], 1]))
abs_pos_K = model.abs_pos_K_emb_dropout(model.abs_pos_K_emb(positions))
abs_pos_V = model.abs_pos_V_emb_dropout(model.abs_pos_V_emb(positions))

time_matrix_K = model.time_matrix_K_dropout(model.time_matrix_K_emb(torch.LongTensor(time_matrix)))
time_matrix_V = model.time_matrix_V_dropout(model.time_matrix_V_emb(torch.LongTensor(time_matrix)))

time_matrix_c_K = model.time_matrix_c_K_dropout(model.time_matrix_c_K_emb(torch.LongTensor(time_matrix_c)))
time_matrix_c_V = model.time_matrix_c_V_dropout(model.time_matrix_c_V_emb(torch.LongTensor(time_matrix_c)))

timeline_mask = torch.BoolTensor(log_seqs == 0)
seqs *= ~timeline_mask.unsqueeze(-1) # broadcast in last dim

tl = seqs.shape[1] # time dim len for enforce causality
attention_mask = ~torch.tril(torch.ones((tl, tl), dtype=torch.bool))

same_time_mask = (time_matrix == 0)

queries = model.attention_layernorms[0](seqs)
```

```python
# Self-attention, Q=layernorm(seqs), K=V=seqs
# seqs = torch.transpose(seqs, 0, 1) # (N, T, C) -> (T, N, C)
mha_outputs = model.attention_layers[0](queries, seqs,
                                timeline_mask, attention_mask, same_time_mask,
                                time_matrix_K, time_matrix_V,
                                abs_pos_K, abs_pos_V,
                                time_matrix_c_K, time_matrix_c_V
                                       )
seqs = queries + mha_outputs
# seqs = torch.transpose(seqs, 0, 1) # (T, N, C) -> (N, T, C)

# Point-wise Feed-forward, actually 2 Conv1D for channel wise fusion
seqs = model.forward_layernorms[0](seqs)
seqs = model.forward_layers[0](seqs)
seqs *=  ~timeline_mask.unsqueeze(-1)
```

```python
Q, K, V = model.attention_layers[1].Q_w(queries), model.attention_layers[1].K_w(seqs), model.attention_layers[1].V_w(seqs)

# head dim * batch dim for parallelization (h*N, T, C/h)
Q_ = torch.cat(torch.split(Q, model.attention_layers[1].head_size, dim=2), dim=0)
K_ = torch.cat(torch.split(K, model.attention_layers[1].head_size, dim=2), dim=0)
V_ = torch.cat(torch.split(V, model.attention_layers[1].head_size, dim=2), dim=0)

time_matrix_K_ = torch.cat(torch.split(time_matrix_K, model.attention_layers[1].head_size, dim=3), dim=0)
time_matrix_V_ = torch.cat(torch.split(time_matrix_V, model.attention_layers[1].head_size, dim=3), dim=0)
abs_pos_K_ = torch.cat(torch.split(abs_pos_K, model.attention_layers[1].head_size, dim=2), dim=0)
abs_pos_V_ = torch.cat(torch.split(abs_pos_V, model.attention_layers[1].head_size, dim=2), dim=0)
time_matrix_c_K_ = torch.cat(torch.split(time_matrix_c_K, model.attention_layers[1].head_size, dim=3), dim=0)
time_matrix_c_V_ = torch.cat(torch.split(time_matrix_c_V, model.attention_layers[1].head_size, dim=3), dim=0)

# batched channel wise matmul to gen attention weights
attn_weights = Q_.matmul(torch.transpose(K_, 1, 2))
attn_weights += Q_.matmul(torch.transpose(abs_pos_K_, 1, 2))
attn_weights += time_matrix_K_.matmul(Q_.unsqueeze(-1)).squeeze(-1)
attn_weights += time_matrix_c_K_.matmul(Q_.unsqueeze(-1)).squeeze(-1)

# seq length adaptive scaling
attn_weights = attn_weights / (K_.shape[-1] ** 0.5)

# fixed a bug pointed out in https://github.com/pmixer/TiSASRec.pytorch/issues/2
# timeline_mask = timeline_mask.unsqueeze(-1).expand(attn_weights.shape[0], -1, attn_weights.shape[-1])
timeline_mask = timeline_mask.unsqueeze(-1).repeat(model.attention_layers[1].head_num, 1, 1)
timeline_mask = timeline_mask.expand(-1, -1, attn_weights.shape[-1])
attention_mask = attention_mask.unsqueeze(0).expand(attn_weights.shape[0], -1, -1)
paddings = torch.ones(attn_weights.shape) *  (-2**32+1) # -1e23 # float('-inf')
paddings = paddings.to(model.attention_layers[1].dev)
same_time_mask = same_time_mask.repeat(model.attention_layers[1].head_num, 1, 1)
attn_weights = torch.where(timeline_mask, paddings, attn_weights) # True:pick padding
attn_weights = torch.where(attention_mask, paddings, attn_weights) # enforcing causality
attn_weights = torch.where(same_time_mask, paddings, attn_weights)

attn_weights = model.attention_layers[1].softmax(attn_weights) 
```

```python
mask = np.triu(np.ones_like(attn_weights[0].detach().numpy()))
```

```python
seqs = model.item_emb(torch.LongTensor(log_seqs).to(model.dev))
seqs *= model.item_emb.embedding_dim ** 0.5
seqs = model.item_emb_dropout(seqs)

positions = np.tile(np.array(range(log_seqs.shape[1])), [log_seqs.shape[0], 1])
positions = torch.LongTensor(positions).to(model.dev)
abs_pos_K = model.abs_pos_K_emb(positions)
abs_pos_V = model.abs_pos_V_emb(positions)
abs_pos_K = model.abs_pos_K_emb_dropout(abs_pos_K)
abs_pos_V = model.abs_pos_V_emb_dropout(abs_pos_V)

time_matrix = torch.LongTensor(time_matrix).to(model.dev)
time_matrix_K = model.time_matrix_K_emb(time_matrix)
time_matrix_V = model.time_matrix_V_emb(time_matrix)
time_matrix_K = model.time_matrix_K_dropout(time_matrix_K)
time_matrix_V = model.time_matrix_V_dropout(time_matrix_V)

time_matrix_c = torch.LongTensor(time_matrix_c).to(model.dev)
time_matrix_c_K = model.time_matrix_c_K_emb(time_matrix_c)
time_matrix_c_V = model.time_matrix_c_V_emb(time_matrix_c)
time_matrix_c_K = model.time_matrix_c_K_dropout(time_matrix_c_K)
time_matrix_c_V = model.time_matrix_c_V_dropout(time_matrix_c_V)

# mask 0th items(placeholder for dry-run) in log_seqs
# would be easier if 0th item could be an exception for training
timeline_mask = torch.BoolTensor(log_seqs == 0).to(model.dev)
seqs *= ~timeline_mask.unsqueeze(-1) # broadcast in last dim

feature_table = []
if (model.args.model.aux_info.clac_hlv_nm == True):
    clac_hlv_nm_Q_ = model.clac_hlv_nm_Q_emb(torch.LongTensor(clac_hlv_nm).to(model.dev))
    clac_hlv_nm_K_ = model.clac_hlv_nm_K_emb(torch.LongTensor(clac_hlv_nm).to(model.dev))
    clac_hlv_nm_Q_ = model.clac_hlv_nm_Q_dropout(clac_hlv_nm_Q_)
    clac_hlv_nm_K_ = model.clac_hlv_nm_K_dropout(clac_hlv_nm_K_)
    clac_hlv_nm_attr = torch.stack([clac_hlv_nm_Q_, clac_hlv_nm_K_])
    feature_table.append(clac_hlv_nm_attr)

if (model.args.model.aux_info.clac_mcls_nm == True):
    clac_mcls_nm_Q_ = model.clac_mcls_nm_Q_emb(torch.LongTensor(clac_mcls_nm).to(model.dev))
    clac_mcls_nm_K_ = model.clac_mcls_nm_K_emb(torch.LongTensor(clac_mcls_nm).to(model.dev))
    clac_mcls_nm_Q_ = model.clac_mcls_nm_Q_dropout(clac_mcls_nm_Q_)
    clac_mcls_nm_K_ = model.clac_mcls_nm_K_dropout(clac_mcls_nm_K_)
    clac_mcls_nm_attr = torch.stack([clac_mcls_nm_Q_, clac_mcls_nm_K_])
    feature_table.append(clac_mcls_nm_attr)

if (model.args.model.aux_info.cop_c == True):
    cop_c_Q_ = model.cop_c_Q_emb(torch.LongTensor(cop_c).to(model.dev))
    cop_c_K_ = model.cop_c_K_emb(torch.LongTensor(cop_c).to(model.dev))
    cop_c_Q_ = model.cop_c_Q_dropout(cop_c_Q_)
    cop_c_K_ = model.cop_c_K_dropout(cop_c_K_)
    cop_c_attr = torch.stack([cop_c_Q_, cop_c_K_])
    feature_table.append(cop_c_attr)

if (model.args.model.aux_info.de_dt_month == True):
    de_dt_month_Q_ = model.de_dt_month_Q_emb(torch.LongTensor(de_dt_month).to(model.dev))
    de_dt_month_K_ = model.de_dt_month_K_emb(torch.LongTensor(de_dt_month).to(model.dev))
    de_dt_month_Q_ = model.de_dt_month_Q_dropout(de_dt_month_Q_)
    de_dt_month_K_ = model.de_dt_month_K_dropout(de_dt_month_K_)
    de_dt_month_attr = torch.stack([de_dt_month_Q_, de_dt_month_K_])
    feature_table.append(de_dt_month_attr)

if (model.args.model.aux_info.buy_am == True) & (model.args.model.aux_info.chnl_dv == True):
    am_chnl = np.concatenate([buy_am.reshape(-1, model.maxlen, 1), chnl_dv.reshape(-1, model.maxlen, 1)], axis=-1)
elif (model.args.model.aux_info.buy_am == True) & (model.args.model.aux_info.chnl_dv == False):
    am_chnl = buy_am.reshape(-1, model.maxlen, 1)
else:
    am_chnl = chnl_dv.reshape(-1, model.maxlen, 1)
if not (model.args.model.aux_info.buy_am == False) & (model.args.model.aux_info.chnl_dv == False):
    am_chnl_Q_ = model.buy_am_n_chnl_dv_Q(torch.FloatTensor(am_chnl).to(model.dev))
    am_chnl_K_ = model.buy_am_n_chnl_dv_K(torch.FloatTensor(am_chnl).to(model.dev))
    am_chnl_Q_ = model.buy_am_n_chnl_dv_Q_dropout(am_chnl_Q_)
    am_chnl_K_ = model.buy_am_n_chnl_dv_K_dropout(am_chnl_K_)
    am_chnl_attr = torch.stack([am_chnl_Q_, am_chnl_K_]).view((2, -1, model.args.model.args.maxlen, model.args.model.args.seq_attr_hidden_units))
    feature_table.append(am_chnl_attr)

if model.seq_feat_num != 0:
    feature_table = torch.stack(feature_table)
else:
    feature_table = None

same_time_mask = model.get_same_time_mask(time_matrix)
extended_attention_mask = model.get_attention_mask(torch.LongTensor(log_seqs).to(model.dev))
```

```python
output1 = model.trm_encoder.layer[0](seqs, feature_table, abs_pos_K, abs_pos_V, time_matrix_K, time_matrix_V, time_matrix_c_K, time_matrix_c_V,
                    extended_attention_mask, same_time_mask, timeline_mask)

res = []
for idx, f in enumerate(feature_table):
    q = model.trm_encoder.layer[1].multi_head_attention.transpose_for_scores_attribute(model.trm_encoder.layer[1].multi_head_attention.query_layers[idx](f[0]), idx)
    k = model.trm_encoder.layer[1].multi_head_attention.transpose_for_scores_attribute(model.trm_encoder.layer[1].multi_head_attention.key_layers[idx](f[1]), idx)
    attention_scores = torch.matmul(q, k.transpose(-1, -2))
    attention_scores = attention_scores/math.sqrt(2)
    res.append(attention_scores)
attribute_attention_table = torch.cat(res,dim=-2)
attention_scores = torch.sum(attribute_attention_table, dim=-2)
```

```python
weights_ = attribute_attention_table / attention_scores.unsqueeze(-2)

data1 = weights_[:,:,:110,:].mean().detach().numpy()
data2 = weights_[:,:,110:220,:].mean().detach().numpy()
data3 = weights_[:,:,220:,:].mean().detach().numpy()

plt.bar(['계열사', '구매 월', '구매 금액, 온라인/오프라인 여부'] ,[data1, data2, data3], color='darkblue')
for idx, x in enumerate([data1, data2, data3]):
    if x > 0:
        plt.text(idx, x, x, horizontalalignment='center', verticalalignment='bottom')
    else:
        plt.text(idx, x, x, horizontalalignment='center', verticalalignment='top')
plt.title('Side Information별 합산 비중')
plt.show()
```
